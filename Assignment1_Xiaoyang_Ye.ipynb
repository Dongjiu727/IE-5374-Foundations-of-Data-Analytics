```python
#1.Computation of inter-event time
import numpy as np
def cal_inter_event_time(even_size_list,min_size): 
  inter_time = [] 
  pre_event_time = 0
  for j in range(0, len(even_size_list)):
    if even_size_list[j] >= min_size:
      if j - pre_event_time != 0:
        inter_time.append(j - pre_event_time)
      pre_event_time = j
  return np.array(inter_time)


even_size_list_1 = [10,1,6,4,3,2,7,0,8,8,2,0,7,7,1,0]
cal_inter_event_time (even_size_list_1,5)

```

```python
import pandas as pd
import numpy as np
df_nursing = pd.read_csv('/content/COVID-19 Nursing Home Data 09.11.2022.csv')
df_copy = df_nursing.copy()
print("Task 2-4: ") 
# Make a new column to display the facilities with (only) 100% bed occupancy, which will be "Yes"
df_fullBed = df_copy.assign(
    fulloccupied = np.where(df_copy['Total Number of Occupied Beds'] >= df_copy['Number of All Beds'], "Yes","NO") 
)

# Sellect the facilities in Middlesex county in MA with 100% bed occupancy
pd_result = df_fullBed.loc[(df_fullBed['Provider State'] == "MA") & (df_fullBed['County'] == "Middlesex")& (df_fullBed['fulloccupied'] == "Yes"),["Provider Name", "County"]]

# Make a new column "count" to count the weeks with 100% bed occupancy
pd_result_1 = pd_result.groupby(pd_result['Provider Name']).agg(
    count = pd.NamedAgg(column = "County", aggfunc = "count")).reset_index()

# Count the number of facilities
display((pd_result_1["count"] >= 3).sum())
```
